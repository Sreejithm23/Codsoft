```python
df.dropna(subset=['Rating'],inplace=True)
df.isna().sum()
```

```python
df.shape
```

```python
df.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre','Duration'],inplace=True)
df.isna().sum()
```

```python
df.shape
```

```python
df.drop(columns=['Name','Duration'],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df
```

```python
df['Votes']=df['Votes'].str.replace(',','')
df['Votes']=df['Votes'].str.replace('.','')
df['Votes']=df['Votes'].astype(int)
df['Year']=df['Year'].str.strip('()').astype(int)
df.dtypes

```

```python
sns.histplot(df["Rating"], kde=True, stat="density")
plt.title('Distribution of rating')
plt.show()
```

```python
#plotting top ten high average rated Genre
genre_rate=df.groupby('Genre').agg({'Rating':'mean'}).sort_values(by='Rating',ascending=False).reset_index().head(10)
genre=sns.barplot(x='Rating',y='Genre',data=genre_rate)
genre.bar_label(genre.containers[0])
plt.title("Top 10 high average rated genre")
plt.show()
```

```python
#plotting top ten high average rated Directors
dr_rate=df.groupby('Director').agg({'Rating':'mean'}).sort_values(by='Rating',ascending=False).reset_index().head(10)
dr=sns.barplot(x='Rating',y='Director',data=dr_rate)
dr.bar_label(dr.containers[0])
plt.title("Top 10 high average rated directors")
plt.show()
```

```python
#plotting top ten high average rated Actors
actor1_rate=df.groupby('Actor 1').agg({'Rating':'mean'}).sort_values(by='Rating',ascending=False).reset_index().head(10)
ar=sns.barplot(x='Rating',y='Actor 1',data=actor1_rate)
ar.bar_label(ar.containers[0])
plt.title("Top 10 high average rated actors")
plt.show()
```

```python
df['Genre_encoded']=round(df.groupby('Genre')['Rating'].transform('mean'),1)
df['Director_encoded']=round(df.groupby('Director')['Rating'].transform('mean'),1)
df['Actor1_encoded']=round(df.groupby('Actor 1')['Rating'].transform('mean'),1)
df['Actor2_encoded']=round(df.groupby('Actor 2')['Rating'].transform('mean'),1)
df['Actor3_encoded']=round(df.groupby('Actor 3')['Rating'].transform('mean'),1)
df.drop(columns=['Genre','Director','Actor 1','Actor 2','Actor 3'],inplace=True)
df.head()
```

```python
plt.figure(figsize=(6,5))
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,cmap='magma',annot=True)
plt.title('Correlation')
plt.show()
```

```python
x=df.drop('Rating',axis=1)
y=df['Rating']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
y_pred_lr=lr_model.predict(x_test)
```

```python
errors=abs(y_pred_lr-y_test)
mape_lr=100*(errors/y_test)
accuracy_lr=100-np.mean(mape_lr)
print("Accuracy Score of the model is",round(accuracy_lr,2),'%')
```

```python
results=pd.DataFrame({'actual':y_test,'predicted':y_pred_lr.ravel(),'residual':y_test-y_pred_lr})
results
```

```python
sns.scatterplot(x=results['actual'],y=results['predicted'])
plt.title('Actual vs predicted')
plt.show()
```

```python
sns.scatterplot(x=results['predicted'],y=results['residual'])
plt.axhline(0, c='red')
plt.title('scatterplot of residuals over predicted')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
dt_model.fit(x_train,y_train)
y_pred_dt=dt_model.predict(x_test)
```

```python
errors=abs(y_pred_dt-y_test)
mape_dt=100*(errors/y_test)
accuracy_dt=100-np.mean(mape_dt)
print('Accuracy Score of the model is',round(accuracy_dt,2),'%')
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor()
rf_model.fit(x_train,y_train)
y_pred_rf=rf_model.predict(x_test)
```

```python
errors=abs(y_pred_rf-y_test)
mape_rf=100*(errors/y_test)
accuracy_rf=100-np.mean(mape_rf)
print('Accuracy Score of the model is',round(accuracy_rf,2),'%')
```
