```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/advertising.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='r')
```

```python
sns.regplot(x=df['Radio'],y=y,color='g')
```

```python
sns.regplot(x=df['Newspaper'],y=y,color='b')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=9)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
print(model.predict([[243.7,121.8,134.8]]))
```

```python
print("The slopes are",model.coef_)
```

```python
print("The constant is",model.intercept_)
```

```python
lst=['TV','Radio','Newspaper']
k=0
for i in lst:
  if(k<3):
    print("The slope for",i,"is",model.coef_[k])
  k+=1
```

```python
print("The slopes are")
list(zip(x,model.coef_))
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("MAE is",mean_absolute_error(y_test,y_pred))
print("MAPE is",mean_absolute_percentage_error(y_test,y_pred))
print("MSE is",mean_squared_error(y_test,y_pred))
mse=mean_squared_error(y_test,y_pred)
print("RMSE is",np.sqrt(mse))
print("R2 Score is",r2_score(y_test,y_pred))
```
